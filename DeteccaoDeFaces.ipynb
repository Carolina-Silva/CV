```python
import dlib
from google.colab.patches import cv2_imshow
import cv2
```

```python
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/CV/recusos/teste.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
```

```python
detector_face_hog = dlib.get_frontal_face_detector()
```

```python
detect = detector_face_hog(img, 1)
```

```python
detect, len(detect)
```

```python
for face in detect:
  # print(face)
  # print(face.left())
  # print(face.top())
  # print(face.right())
  # print(face.bottom())

  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(img, (l,t), (r,b), (0,255,255) , 2)
cv2_imshow(img)
```
